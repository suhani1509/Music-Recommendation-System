```python
from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware
import pickle
import pandas as pd

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_methods=["*"],
    allow_headers=["*"],
)

# Load karte waqt variable name 'cosine_mat' rakh sakte hain
music_df = pickle.load(open('music_list.pkl', 'rb'))
cosine_mat = pickle.load(open('similarity.pkl', 'rb')) # File se load kiya

@app.get("/recommend")
def get_recommend(song: str):
    try:
        # 1. Index dhundna
        idx = music_df[music_df['song'].str.lower() == song.lower()].index[0]
        
        # 2. 'cosine_mat' ka use karke distances nikalna
        distances = sorted(list(enumerate(cosine_mat[idx])), reverse=True, key=lambda x: x[1])
        
        recs = []
        for i in distances[1:6]:
            recs.append({
                "artist": music_df.iloc[i[0]].artist,
                "song": music_df.iloc[i[0]].song
            })
        return {"recommendations": recs}
    except Exception as e:
        return {"recommendations": [], "error": str(e)}
```
